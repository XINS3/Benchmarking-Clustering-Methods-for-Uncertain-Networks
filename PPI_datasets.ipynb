```python
def read_clustering(file):
    import csv
    cluster=[]
    with open (file, 'r' )as f:
        for line in f:
            nodes=line.strip().split()
            cluster.append(set(g.upper().strip() for g in nodes))
    return cluster

# test
#file='//Users//xinsh758//Documents//academic//probabilistic_networks//data/proteins//cl1_benchmark_data//datasets//krogan2006_core.txt'
file='//Users//xinsh758//Documents//academic//probabilistic_networks//data/proteins//cl1_benchmark_data//gold_standard//mips_3_100.txt'
c=read_clustering(file)
c
    
```

```python
def read_network(file):
    edges=[]
    with open(file,'r') as f:
        for line in f:
            parts=line.strip().split('\t')
            v, u, p=parts
            edges.append((u,v,float(p)))
    return edges


# text
file='//Users//xinsh758//Documents//academic//probabilistic_networks//data/proteins//cl1_benchmark_data//datasets//krogan2006_core.txt'
e=read_network(file)
e
```

```python
def intersects_node(clusters, edges):
    # intersection node
    #cluster_=[set(cluster_string.split()) for group in clusters for cluster_string in group]
    cluster_nodes=set().union(*clusters)
    network_nodes=set(u for u,_,_ in edges).union(v for _,v, _ in edges)
    valid_node=cluster_nodes & network_nodes
    # remove valid nodes that only connect to invalide node
    connect_valid=set()
    for u,v,_ in edges:
        if u in valid_node and v in valid_node:
            connect_valid.add(u)
            connect_valid.add(v)

    return valid_node & connect_valid

# text
n=intersects_node(c,e)
len(set(n))

```

```python
def map_name_ID(intersec_node):
    return {node: str(i) for i, node in enumerate(sorted(intersec_node))}
mapID=map_name_ID(n)
```

```python
def remap_clustering(cluster_nodes, intersec_node, mapID,filename):
    remap=[]
    for c in cluster_nodes:
       
        filtered=c & intersec_node
   
        if filtered:
            remap.append([mapID[name] for name in filtered])
    with open(filename,'w') as f:
        for c in remap:
            f.write(' '.join(c) +'\n')
    


#re_clus=remap_clustering(c,n,mapID)

#re_clus
```

```python
def remap_network(edge,mapID,intersec_node,filename):
    net= [
        (mapID[u],mapID[v],p)
        for u,v ,p in edge
        if u in intersec_node and v in intersec_node
    ]
    with open(filename, 'w') as f:
        for u, v, w in net:
            f.write(f"{u} {v} {w:.3f}\n")


#remap_network(e,mapID,n)
```
