## F1-score

```python
from itertools import combinations

def generate_pairs(clusters):
    """
    Generate all unique item pairs within each cluster.
    Args:
        clusters (dict): Dictionary mapping cluster_id -> set of items
    Returns:
        Set of tuples: Each tuple is a sorted pair (item1, item2)
    """
    pairs = set()
    
    # If it's a dict, get the values
    cluster_list = clusters.values() if isinstance(clusters, dict) else clusters

    for items in cluster_list:
        for pair in combinations(sorted(items), 2):
            
            pairs.add(pair)
    return pairs

def pairwise_f1_score(predicted_clusters, true_clusters):
    """
    Compute pairwise precision, recall, and F1-score between predicted and true clusterings.
    
    Args:
        predicted_clusters (dict): cluster_id -> set of items (can be overlapping)
        true_clusters (dict): cluster_id -> set of items (usually non-overlapping)
    
    Returns:
        dict: precision, recall, f1 score, and pair counts
    """
    predicted_pairs = generate_pairs(predicted_clusters)
    true_pairs = generate_pairs(true_clusters)

    # print(predicted_pairs)
    # print(true_pairs)
    correct_pairs = predicted_pairs & true_pairs
    #print(correct_pairs)
    
    precision = len(correct_pairs) / len(predicted_pairs) if predicted_pairs else 0
    recall = len(correct_pairs) / len(true_pairs) if true_pairs else 0
    f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0
    
    return {
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'correct_pairs': len(correct_pairs),
        'predicted_pairs': len(predicted_pairs),
        'true_pairs': len(true_pairs)
    }


predicted=[[1,2,3,5,6,9],[4,5]]
true=[[1,2],[3,4]]
results = pairwise_f1_score(predicted, true)
print(results)
```
