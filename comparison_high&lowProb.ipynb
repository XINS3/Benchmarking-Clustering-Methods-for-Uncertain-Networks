```python
# low prob
#------------------


# exported json reading, os ordering, expected mod
import json
import Expected_mod as expmod
from tkinter import Tcl
import os
import load_graph

# ordering graph datasets based on k
dataset_path='mcp_acp_data//l10_p0.3_evolving_lowP1//dataset//'
filelists=[i for i in os.listdir(dataset_path)]
filelists=Tcl().call('lsort','-dict',filelists)
print(filelists)

# go to result path
result_path='mcp_acp_data//l10_p0.3_evolving_lowP1//result//'

# go through file one by one - loop
# extract filename and store it
# go into file, read json file, return a dic
algorithms=[i for i in os.listdir(result_path) if i != '.DS_Store']
print(algorithms)
result={}
for alg in algorithms:
    alg_name=alg
    alg_result_path=result_path+alg_name+'//'
    #print(alg_result_path)
    file=[i for i in os.listdir(alg_result_path) if i[-5:]=='.json'][0]
    #print(file)
    with open(alg_result_path+file,'r') as fp:
        clusterings=json.load(fp)
    #print(data)
    print(alg_name)
    dic={}
# go through dic with k
    for k in range(2,11):

# use k as index in filelists to return graph
        graph=filelists[k-2]

# extract n,e, p from networkx type graph
        g=load_graph.read_g(dataset_path+graph)
        edge=[]
        p=[]
        for u,v,w in g.edges(data=True):
            edge.append((u,v))
            p.append(w['weight'])
# use k as key from dic return clustering
       
        clustering=clusterings[str(k)]
# calculate expected value
# store name.update({k:value})
        expectedValue=expmod.APWP(edge,p,clustering)
        #print(expectedValue)
        dic.update({k:expectedValue})
        print(graph,': ','edge: ',len(edge))
    print(alg_name, ' :',dic)
    result[alg_name]=dic
    
```
